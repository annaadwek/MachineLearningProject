## Normulaizing the Data

Normalization standardised the range of values. If a feature’s variance order of magnitude is more than the variance of other features, that particular feature might dominate other features in the dataset and cause problems (VINAYVIKRAM, 2020). This effects algorithms that use use Eucledian distance between two data points (VINAYVIKRAM, 2020). The random forest model should not be as affected as other models since tree based models are not distance based models and can handle varying ranges of features (VINAYVIKRAM, 2020). Though scaling the data for a random forest may not affect results as significantly as other algorithms, it is always recombined to implement standardization (Fernandez, 2021).

# Splitting the Dataset

```python
#Importing package
from sklearn.model_selection import StratifiedKFold
```

```python
# fixing the randomness
import random
random.seed(42)
np.random.seed(42)
%matplotlib inline
```

```python
#seperating the X and the Y from the dataset

x = Dataset_norm.drop(labels='isFraud', axis=1) # Features
y = Dataset_norm["isFraud"]               # Response

print(x.shape)
print(y.shape)
#getting just the values for the sake of processing (its a numpy array with no columns)
x=x.to_numpy()
y=y.to_numpy()
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show()
```

```python
# shuffle the dataset
# keep the original data, by creating new arrays, x_s, and y_s
rng = np.random.default_rng(42)

shuffle = rng.permutation(len(y))
x_s = x[shuffle]
y_s = y[shuffle]
```

```python
# create a stratified splitter from scikit-learn

# it will split the dataset in 2 part, keeping the same proportion of each
# class in each split
skf = StratifiedKFold(n_splits=2)
# the index of both dataset
train_index_0,  test_index = [i[0] for i in skf.split(x_s, y_s)]
```

```python
# compute which class has the lowest occurence in the train dataset
lowest = np.min(np.unique(y_s[train_index_0], return_counts=True)[1])
```

```python
# cut all the other class to the number of occurence of the least present class
# in the training dataset

# if there's 400 blue and 100 red, after this there will be 100 blue and 100 red
train_index = train_index_0[np.concatenate(
    [np.where(y_s[train_index_0] == i)[0][:lowest]
     for i in np.unique(y_s)])]
```

```python
plt.scatter(x[train_index][:, 0], x[train_index][:, 1], c=y[train_index],
            label='train')
plt.scatter(x[test_index][:, 0], x[test_index][:, 1], c=y[test_index],
            cmap='cool', label='test', marker='*')
plt.legend()
plt.show()
```

# Implementing Random Forest

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing the random forest classifier algorithm function from sklearn
from sklearn.ensemble import RandomForestClassifier
# the random forest is stochastic, so we use a random_state parameter to fix the result
clf = RandomForestClassifier(random_state=42)
```

```python
# random forest model creation
clf.fit(x[train_index], y[train_index])
clf.score(x[test_index], y[test_index])
```

```python
clf2 = RandomForestClassifier(max_features=5, random_state=42)
```

```python
clf2.fit(x[train_index], y[train_index])
clf2.score(x[test_index], y[test_index])
```

```python
clfm = RandomForestClassifier(
    random_state=42, n_estimators=100, max_samples=0.5, 
    max_features='sqrt', max_depth=15)
```

```python
clfm.fit(x[train_index], y[train_index])
res_p = clfm.predict_proba(x[test_index])
res = clfm.predict(x[test_index])
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, \
    recall_score, precision_score, ConfusionMatrixDisplay, RocCurveDisplay
print('acc:', accuracy_score(y[test_index], res))
print('ppv:', precision_score(y[test_index], res))
print('tpr:', recall_score(y[test_index], res))
print('f1 :', f1_score(y[test_index], res))
print('auc :', roc_auc_score(y[test_index], res_p[:, 1]))
```
